```r
conc = c(0.03125,0.0625,0.125,0.25,0.5,1)
```

```r
freq = c(21,6,8,19,17,3)
```

```r
# Geometric mean
logs=sum(freq*log10(conc))/sum(freq)
geo_mean=10^logs
geo_mean
```
